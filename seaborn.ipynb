```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np  
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset("penguins")
penguins.head(5)
```

```python
penguins["species"].value_counts()
```

```python
penguins["island"].value_counts()
```

```python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
plt.show()
```

```python
sns.set_context("notebook")

sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", palette="Dark2")
sns.despine(left=True)
plt.show()
```

```python
sns.scatterplot(data = penguins, x ="species", hue ="island", y = "body_mass_g", style = "sex", alpha = 0.5)
```

```python
sns.stripplot(data = penguins, x = "species", y = "body_mass_g", hue = "island", dodge = "True", jitter=True)
```

```python
# Swarm Plot
sns.set_context("notebook")
sns.swarmplot(data = penguins, x = "species", y = "body_mass_g", hue = "island")
sns.despine()
```

```python
#histogram
sns.histplot(data = penguins, x = "body_mass_g", hue = "sex", multiple="stack"  )
```

```python
# regression plot:
sns.regplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", color="green")
```

```python
#line plot
sns.lineplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", hue="island", style="sex")
```

```python
# jointplots
sns.jointplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm" )
```

```python
sns.jointplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", hue="sex")
```

```python
sns.jointplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", kind="reg")
```

```python
# bar plot

sns.barplot(data = penguins,x = "species", y = "body_mass_g", hue ="sex", palette=["red", "blue"], estimator=np.var)
```

```python
# countplot:
sns.countplot(data = penguins, x ="species", hue= "island")   
```

```python
# boxplot
sns.boxplot(data = penguins,x= "species", y = "body_mass_g", hue = "sex")
```

```python
# violin plot

sns.violinplot(data = penguins, x= "species", y = "body_mass_g", hue="sex", split="True", inner= 'box')
```

```python
# width of the violin : density of the data at diffrent value
# middle line : median
# thick bar : interquartile range ( if inner = 'box')
```

```python
sns.violinplot(data = penguins, x= "species", y = "body_mass_g", palette="pastel")
sns.swarmplot(data = penguins, x = "species", y = "body_mass_g", color="black", hue= "island",size= 3, palette="dark")
```

```python
# KDE -> for estimating hte probability density function of continuous or non-parametric data
# Or in simple term : a smooth curve that show the distribution of data - like a smoothed - out version
# it helps you see where the data is concentrated (peaks) and how it spreads

sns.kdeplot(data = penguins, x = "body_mass_g", hue = "species", fill = True)
```

```python
#Heatmap : a 2D color-coded matrix that shows relationships or patterns between two variables.
# It's commonly used to :
#   - Show correlations between features
#   - visualize confusion matrices -> will see in the later stages
#   - highliht missing data -> sns.heatmap(df.isnull())
```

```python
# Heatmap

penguins.head()
```

```python
coloums =["bill_length_mm",	"bill_depth_mm",	"flipper_length_mm","body_mass_g"]
penguins[coloums].corr()
```

```python
sns.heatmap(data = penguins[coloums].corr())
```

```python
sns.heatmap(data = penguins[coloums].corr(), annot=True, cmap="Blues", vmin= -0.2)
```

```python
# Rug plot:

sns.rugplot(data = penguins, y="body_mass_g", hue = "species", palette= "Set2", height= 0.2)
```

```python
# pair plot:

sns.pairplot(data= penguins)
```

```python
# pair plot:

sns.pairplot(data= penguins, hue = "sex", palette= "Set2", diag_kind= "kde")
```

```python
# pair grid
graph = sns.PairGrid(data= penguins, hue= "sex", palette= "Set2")
graph.map_upper(sns.scatterplot) 
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
```
